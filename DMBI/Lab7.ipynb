```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Example Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 45, 56, 78]

# Line Plot
plt.figure(figsize=(8, 5))
plt.plot(categories, values, marker='o', color='b', label='Line Plot')
plt.title('Line Plot Example', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='c', alpha=0.7)
plt.title('Bar Chart Example', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.show()

# Scatter Plot
plt.scatter(categories, values, color='blue', marker='o')
plt.title('Scatter Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Pie Chart
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
